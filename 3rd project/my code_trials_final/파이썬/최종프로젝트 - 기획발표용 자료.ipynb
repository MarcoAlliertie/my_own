```python
#최종프로젝트용 자료 - 전체 교통사고 및 자전거 사고 통계 
#출처 - https://taas.koroad.or.kr/sta/acs/gus/selectTfcacdOccrrncSttusAnnual.do?menuId=WEB_KMP_OV

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

year = []
for i in range(2018,2023):
    year.append(i)
    

#전체 교통사고
acc = [217148, 229600, 209654, 203130, 196836] #2018년부터 발생한 총 교통사고 건수
dead = [3781, 3349, 3081, 2916, 2735 ] #연도별 사망자
injured = [323037, 341712, 306194, 291608, 281803] #연도별 부상자

#자전거 교통사고
bi_acc = [4771, 5633, 5667, 5509, 5393]
bi_dead = [91, 79, 83, 70, 91]
bi_inj = [5041, 6020, 6150,5999,5856 ]




stat_ba = []
stat_bd = []
stat_bi = []
for a, d, i, ba, bd ,bi in zip(acc,dead,injured,bi_acc,bi_dead,bi_inj):
    stat_ba.append(round(ba/a*100,2))
    stat_bd.append(round(bd/d*100,2))
    stat_bi.append(round(bi/i*100,2))

print('2018~2022년 사이 통계')
print('전체 교통사고 대비 자전거 사고(이하 통계1) :', stat_ba)
print('전체 교통사고 사망자 수 대비 자전거 사고 사망자 비율(이하 통계2) :',stat_bd)
print('전체 교통사고 부상자 수 대비 자전거 사고 부상자 비율(이하 통계3) :',stat_bi)
print('통계2 기준 2021년 대비 2022년 증가율(%p) :',(round(stat_bd[4]-stat_bd[3],2)))
print('통계2 기준 2021년 대비 2022년 증가율(%) :',(round((stat_bd[4]-stat_bd[3])/stat_bd[3]*100,2)))

```

```python
plt.xlim(2018,2022)
plt.xticks(range(2018,2023))

plt.title('전체 교통사고(이하 A) 및 자전거 사고(이하 B) 통계')
plt.plot(year,stat_ba, label='A 대비 B(%)')
plt.plot(year,stat_bd, label='A 사망자 대비 B 사망자 비율(%)')
plt.plot(year,stat_bi, label='A 부상자 대비 B 부상자 비율(%)')
plt.grid()

plt.legend(loc='lower right')

plt.xlabel('연도')
plt.ylabel('%')

plt.annotate('', xy=(2021, 2.4), xytext=(2021, 3),
              arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('증가율 38.75%', xy=(2022, 3.33), xytext=(2020.5, 3),
              arrowprops=dict(facecolor='black', shrink=0.05))
plt.savefig('./bicycle_accident.png', dpi =600, bbox_inches = 'tight')
```

- 음성처리 기술의 발전(파이썬 버전)
    - 과거 음성처리는 따른 HMM, GMM 등  통계적, 확률적 기법에 기반하였다. 이 시기의 음성처리 기술은 음성 신호를 기계에 이해시키는 기반을 마련하는 정도에 그쳤다.
    - 2010년대 이래 RNN에서 파생된 LSTM,GRU 등 딥러닝 기술이 음성 처리 기술에 접목되면서 음성 인식, 텍스트 음성 합성, 감정 인식 등 음성 처리 기술의 전 분야에서 성능이 빠르게 진화하였다.
    - 2020년대 들어서는 거의 매년마다 새로운 모델이 등장하였다. 2021년에는 텍스트나 레이블 없이 음성을 합성할 수 있는 HuBERT, 2022년에는 ChatGPT 어플 음성 인식의 기반 기술인 Whisper가 나왔으며 2023년 MS사에서 내놓은 Vall-E는 학습과정에서 배우지 않은 작업을 수행하는 Zero Shot에 따라 단 3초의 음성 샘플만으로 새로운 내용의 음성을 합성할 수 있게 되었다.
-

    - 2010년대 이래 LSTM,GRU 등 딥러닝 기술이 음성 처리 기술에 접목되면서 음성 인식, 텍스트 음성 합성, 감정 인식 등 음성 처리 기술의 전 분야에서 빠르게 진화해옴.

## STP 전략


시장 세분화(Segmentation)
- 패션에 관심이 많거나 지식이 적어 추천이 필요한 소비자층
- 소통의 기회가 적어 대화와 교감이 필요한 소비자층
- 업무특성, 기억력 문제 등으로 인해 일정관리가 필요한 소비자층

표적 시장(Targeting)
- 외로움, 우울 등 부정적인 감정과 그로 인한 악영향에 노출되기 쉬운 소비자층을 주요 표적시장으로 삼고 의사소통 및 감정 교류 기능을 핵심 기능으로 구축.

- 특히 1인 가구의 경우 위와 같은 문제점과 더불어 가내에서 다른 가족 구성원으로부터 도움을 얻기 어려운 환경임을 감안, 이들을 위해 코디 기능과 일정관리 기능을 제공.

포지셔닝(Positioning)
- 다기능에 따른 포지셔닝 : 유사제품들의 파편화된 기능을 통합, 종합적인 솔루션 제공으로 표적시장의 수요에 맞춘 올인원 케어

- 범용성에 따른 포지셔닝 : 특정 연령대에 국한하지 않고 소비자의 인적 사항에 맞춘 쌍방향적 대화 및 스타일링 추천으로 연령대에 의한 시장수요한계를 극복
