````python
def weighted_f1(data,f1_score):
    weighted_f1_score=0
    class_data_count = np.sum(data,axis=0)
    for i in range(len(f1_score)):
        weighted_f1_score += (class_data_count[i]/sum(class_data_count))*f1_score[i]
    return weighted_f1_score
````

````python
# 임베딩 함수 설정
class text_embedding():
    def __init__(self, model_name):
        self.model_name = model_name

    def fit(self, X, y=None):
        return self
  
    def transform(self, X,raw_label):
        embedding_model = SentenceTransformer(self.model_name)
        embedding_vec = embedding_model.encode(raw_label)
        X_embedded = np.concatenate((X, embedding_vec), axis = 1)
        return X_embedded
````

````python
# 음성 모델
def cnn_model(data):
    model=Sequential()
    model.add(InputLayer(shape=(data.shape[1],data.shape[2])))
    model.add(Conv1D(256, kernel_size=5, strides=1, padding='same', activation=LeakyReLU(alpha=0.2)))
    model.add(MaxPooling1D(pool_size=5, strides = 2, padding = 'same'))

    model.add(Conv1D(256, kernel_size=5, strides=1, padding='same', activation=LeakyReLU(alpha=0.2)))
    model.add(MaxPooling1D(pool_size=5, strides = 2, padding = 'same'))

    model.add(Conv1D(128, kernel_size=5, strides=1, padding='same', activation=LeakyReLU(alpha=0.2)))
    model.add(MaxPooling1D(pool_size=5, strides = 2, padding = 'same'))
    model.add(Dropout(0.2))

    model.add(Conv1D(64, kernel_size=5, strides=1, padding='same', activation=LeakyReLU(alpha=0.2)))
    model.add(MaxPooling1D(pool_size=5, strides = 2, padding = 'same'))

    model.add(Flatten())
    model.add(Dense(units=32, activation=LeakyReLU(alpha=0.2)))
    model.add(Dropout(0.3))

    model.add(Dense(units=4, activation='softmax'))
    model.compile(optimizer = 'adamW' , loss = 'categorical_crossentropy' , metrics = ['accuracy','f1_score']) #metrics 추가

    return model
````

````python
def perceptual_sharpness(audio_path, sr=16000, n_fft=2048, hop_length=1024):
    # 음원 파일 로드
    y= audio_path
    sr = sr

    # STFT 수행
    D = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length))

    # 주파수 대역별로 에너지 계산
    energy = np.sum(D, axis=0)

    # 고주파수 대역 성분 추출
    high_freq_energy = energy[3000:8000]  # 예시로 3000Hz에서 6000Hz 사이의 주파수 대역을 고주파수 대역으로 설정

    # Perceptual Sharpness 계산
    sharpness = np.sum(np.log1p(high_freq_energy))

    return sharpness
````

````python
feature_list_1=[]
feature_list_2=[]
for j in tqdm(range(len(glob.glob(train_audio_path+'*.wav')))):
#     for j in tqdm(range(1)): #오류 데이터 추가용

# y, sr = librosa.load('d:/Data2/test/5ebd3dd7c38c123b9ec6deba.wav', sr=16000)
    y, sr = librosa.load(glob.glob(train_audio_path+'*.wav')[j], sr=16000)
     ##Mel-spectrogram 구현
    spectrogram = librosa.stft(y, n_fft=2048, hop_length= 1024) 
    power_spectrogram = spectrogram**2
    mel = librosa.feature.melspectrogram(S=power_spectrogram, sr=sr)
    mel = librosa.power_to_db(np.abs(mel)**2)
    #mfcc 구현
    mfccs = librosa.feature.mfcc(S = mel, n_mfcc=40)
    try:
        mfccs_mean = mfccs.mean(axis=1)
        mfccs_var = mfccs.mean(axis=1)
    except Exception as e:
        print(f'{j}번째 파일에서 문제 발생',e)
    feature_list_1.append(mfccs_mean)
    feature_list_1.append(mfccs_var)
    feature_list_2.append(mfccs_mean)
    feature_list_2.append(mfccs_var)
mfccs  = np.array(feature_list_1).reshape(-1,len(feature_list_1[0])*2)
train_1st_feature = np.hstack((train_1st_feature,mfccs))
train_1st_feature  = np.pad(train_1st_feature,((0,0),(2,2)),mode='constant')
mfccs  = np.array(feature_list_2).reshape(-1,len(feature_list_2[0])*2)
train_2nd_feature = np.hstack((train_2nd_feature,mfccs))
train_2nd_feature = np.pad(train_2nd_feature,((0,0),(2,2)),mode='constant')
````

````python
feature_list=[]
mfccs_list_1=[]
mfccs_list_2 =[]
for j in tqdm(range(len(glob.glob(test_audio_path+'*.wav')))):
#     for j in tqdm(range(1)): #오류 데이터 추가용

# y, sr = librosa.load('d:/Data2/test/5ebd3dd7c38c123b9ec6deba.wav', sr=16000)

    try:
        y, sr = librosa.load(glob.glob(test_audio_path+'*.wav')[j], sr=16000)
        spectrogram = librosa.stft(y, n_fft=2048, hop_length= 1024) 
        power_spectrogram = spectrogram**2
        mel = librosa.feature.melspectrogram(S=power_spectrogram, sr=sr)
        mel = librosa.power_to_db(np.abs(mel)**2)
        #mfcc 구현
        mfccs = librosa.feature.mfcc(S = mel, n_mfcc=40)
        stft = np.abs(spectrogram)
        chroma_stft = librosa.feature.chroma_stft(S=stft,hop_length=1024)
        rms = librosa.feature.rms(y=y)
        spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)
        spectral_bandwidths = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        zero_crossing_rates = librosa.feature.zero_crossing_rate(y=y)
        chroma_cens = librosa.feature.chroma_cens(C=spectrogram, sr=sr)
        tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
        # ps = perceptual_sharpness(audio) #이거 필요없다.
        chroma_stft_mean = chroma_stft.mean()
        chroma_stft_var = chroma_stft.var()
        rms_mean = rms.mean()
        rms_var = rms.var()
        spectral_centroids_mean = spectral_centroids.mean()
        spectral_centroids_var = spectral_centroids.var()
        spectral_bandwidths_mean = spectral_bandwidths.mean()
        spectral_bandwidths_var = spectral_bandwidths.var()
        spectral_rolloff_mean = spectral_rolloff.mean()
        spectral_rolloff_var = spectral_rolloff.var()
        zero_crossing_rates_mean = zero_crossing_rates.mean()
        zero_crossing_rates_var = zero_crossing_rates.var()
        harmony_mean = chroma_cens.mean()
        harmony_var = chroma_cens.var()
        tempo_mean = tempo.mean()
        tempo_var = tempo.var()
        # perceptual_sharpness_mean = ps.mean()
        # perceptual_sharpness_var = ps.var()
        # 피치 추출
        pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    
        # 피치 값이 있는 프레임만 선택
        pitches = pitches[magnitudes > np.median(magnitudes)]
        x = np.arange(len(pitches))
    
        # 1차 함수 근사화 (기울기)
        coeffs_1 = np.polyfit(x, pitches, 1)
        slope = coeffs_1[0]
        #     print("기울기 (1차 함수 근사화):", slope)
    
        # 2차 함수 근사화 (곡률)
        coeffs_2 = np.polyfit(x, pitches, 2)
        curvature = coeffs_2[0]
        #     print("곡률 (2차 함수 근사화):", curvature)
    
        # 3차 함수 근사화 (변곡)
        coeffs_3 = np.polyfit(x, pitches, 3)
        inflexion = coeffs_3[0]
        #     print("변곡 (3차 함수 근사화):", inflexion)
    
        # 피치의 왜도
        pitch_skewness = scipy.stats.skew(pitches)
        #     print("피치의 왜도:", pitch_skewness)
    
        # 0. 피치의 첨도
        pitch_kurtosis = scipy.stats.kurtosis(pitches)
        #     print("피치의 첨도:", pitch_kurtosis)
    
        # 1. 피치의 사분범위
        pitch_iqr = np.percentile(pitches, 75) - np.percentile(pitches, 25)
        #     print("피치의 사분범위:", pitch_iqr)
    
        # 2. 피치의 급격한 변곡 횟수
        pitch_diff = np.diff(pitches)
        sign_changes = np.diff(np.sign(pitch_diff))
        pitch_turning_points = np.count_nonzero(sign_changes != 0)
        #     print("피치의 급격한 변곡 횟수:", pitch_turning_points)
    
        # 3. 피치의 평균과 최소 차이
        mean_pitch = np.mean(pitches)
        min_pitch = np.min(pitches)
        mean_min_diff = mean_pitch - min_pitch
        #     print("피치의 평균과 최소 차이:", mean_min_diff)
    
        # 4. 피치 중 평균이상 값의 평균값
        mean_above_mean_pitch = np.mean(pitches[pitches > mean_pitch])
        #     print("피치 중 평균이상 값의 평균값:", mean_above_mean_pitch)
    
        # 5. 피치의 유성음 구간 변곡의 평균값
        # sign_changes 배열의 길이를 pitches 배열과 맞추기 위해 끝에 False 값을 추가
        sign_changes = np.append(sign_changes, [False, False])
        pitch_turning_values = pitches[sign_changes != 0]
        mean_turning_pitch = np.mean(pitch_turning_values)
        #     print("피치의 유성음 구간 변곡의 평균값:", mean_turning_pitch)
    
        # 6. 피치의 상승 기울기 구간의 중간값
        rising_slopes = pitch_diff[pitch_diff > 0]
        median_rising_slope = np.median(rising_slopes)
        #     print("피치의 상승 기울기 구간의 중간값:", median_rising_slope)
    
        # 7. 피치의 유성음 구간 곡률의 평균값
        curvature_values = np.abs(np.diff(pitch_diff))
        mean_curvature = np.mean(curvature_values)
        #     print("피치의 유성음 구간 곡률의 평균값:", mean_curvature)
    
    
        # 8. 피치의 plateaux at maxima의 사분범위
        # plateaux at maxima 계산
    
        first_derivative = np.diff(pitches, n=1)  # 1차 미분값
        second_derivative = np.diff(pitches, n=2)  # 2차 미분값
        second_derivative= np.append(second_derivative, 0)
    
        ## 1차 미분값 절대값이 가장 작은 순서대로
        top_10_indices = np.argsort(np.abs(first_derivative))[:len(first_derivative)]
    
        # 상위 10개 인덱스 중 2차 미분값이 음수인 인덱스 찾기
        valid_indices = [idx for idx in top_10_indices if second_derivative[idx] < 0]
    
        # 유효한 인덱스의 피치 값 추출
        valid_pitches = pitches[valid_indices]
    
        # 사분범위 계산
        if valid_pitches.size > 0:
            maxima_iqr = np.percentile(valid_pitches, 75) - np.percentile(valid_pitches, 25)
        #     print("피치의 plateaux at maxima의 사분범위:", maxima_iqr)
        else:
            print("plateaux at maxima가 없습니다.")
            maxima_iqr = 0
    
        # 9. 하강 기울기의 사분위수 지속 구간
        max_slope = np.max(np.abs(pitch_diff))
        #     print("피치의 유성음 구간 기울기의 최대값:", max_slope)
    
        # 10. 피치의 유성음 구간 변곡의 표준편차
        falling_slopes = pitch_diff[pitch_diff < 0]
        falling_slope_iqr = np.percentile(falling_slopes, 75) - np.percentile(falling_slopes, 25)
        #     print("피치의 하강 기울기의 사분위수 지속 구간:", falling_slope_iqr)
    
        # 11. 피치의 유성음 구간 변곡의 표준편차
        std_turning_pitch = np.std(pitch_turning_values)
        #     print("피치의 유성음 구간 변곡의 표준편차:", std_turning_pitch)
    except Exception as e:
        print(f'{j}번째 파일에서 문제 발생',e)


        #합치기
    features = np.array([])
#         for t in range(len(mfccs_mean)):
#             features = np.hstack((features,locals()[f'mfccs_mean_{t}'],locals()[f'mfccs_var_{t}']))

    features = np.hstack((features,chroma_stft_mean,chroma_stft_var,rms_mean,rms_var,spectral_centroids_mean
                         , spectral_bandwidths_mean, spectral_rolloff_mean
                         , spectral_rolloff_var,zero_crossing_rates_mean, np.array(harmony_mean), 
                          harmony_var, tempo_var,pitch_kurtosis, pitch_iqr, pitch_turning_points, mean_min_diff, 
mean_above_mean_pitch,  median_rising_slope, mean_curvature, 
maxima_iqr, max_slope,  std_turning_pitch))
    augmenter = naa.NoiseAug() #데이터 증강을 위해 노이즈된 특성 추가
    augmented_features = np.squeeze(augmenter.augment(features))
    features =np.hstack((features,augmented_features))
    mfccs_mean = mfccs.mean(axis=1)
    mfccs_var = mfccs.mean(axis=1)
    mfccs_list_1.append(mfccs_mean)
    mfccs_list_1.append(mfccs_var)
    mfccs_list_2.append(mfccs_mean)
    mfccs_list_2.append(mfccs_var)

    feature_list.append(features)
feature_array  = np.array(feature_list).reshape(-1,len(feature_list[0]))
mfccs  = np.array(mfccs_list_1).reshape(-1,len(mfccs_list_1[0])*2)
test_1st_feature = np.hstack((feature_array,mfccs))
test_1st_feature  = np.pad(test_1st_feature,((0,0),(2,2)),mode='constant')

    

````
